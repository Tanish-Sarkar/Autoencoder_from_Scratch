```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

DATA_PATH = Path("../data/creditcard.csv")
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
class_count = df['Class'].value_counts()
print(class_count)
```

```python
plt.figure(figsize=(5,6))
sns.barplot(x=class_count.index, y=class_count.values, palette='viridis')
plt.xticks([0,1], ["Normal(0)", "Fraud(1)"])
plt.ylabel("Count")
plt.title("Class Distribution")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4)) # (rows, cols, size)

axes[0].hist(df['Time'], bins=50)
axes[0].set_title("Time Distribution")
axes[1].hist(df['Amount'], bins=50)
axes[1].set_title("Amount Distribution")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(30, 34))
corr = df.corr()
sns.heatmap(corr, cmap='viridis', linewidths=0.1, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
